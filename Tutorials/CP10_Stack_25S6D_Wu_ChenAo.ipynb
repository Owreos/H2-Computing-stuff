
## CP10 - Stack 

**Stack** is a Last in First Out (LIFO) structure in which access is completely restricted to just one end – this end is known as top.

### Operations
The basic operations of a stack is to add and remove item from its top. 
* **push()**: Add an item into the stack
* **pop()**: Remove item from the stack

Other supporting functions to be added are:
* **isEmpty()**: Returns true if the stack is empty
* **size()**: Returns the size of the stack
* **peek()**: Peek at the topmost item without removing it

```python
class ArrayStack2(ArrayStack):
    ###  Implementation of ArrayStack2 here
    def __init__(self):
        super().__init__()

    def is_empty(self):
        return not bool(self._items)
    
    def size(self):
        return len(self._items)
    
    def peek(self):
        if not self.is_empty():
            return self._items[-1]
    
    
stack = ArrayStack2()
stack.push('a')
stack.push('b')
print(stack.size())
stack.pop()
print(stack.peek())
stack.pop()
print(stack.is_empty())
print(stack.peek())
```

```python
class ArrayStack3(ArrayStack2):
    ###  Implementation of ArrayStack3 here
    def __init__(self, size: int):
        self.arrsize = size
        self._items = [None]*size
        self.top = -1
    
    def push(self, item):
        if self.top == self.arrsize -1:
            raise Exception("Stack Overflow")
        self.top += 1
        self._items[self.top] = item
    
    def pop(self):
        if self.isEmpty():
            raise Exception("Stack Underflow")
        foo, self._items[self.top] = self._items[self.top], None
        self.top -= 1
        return foo

    def peek(self):
        if self.isEmpty():
            raise Exception("Stack is Empty")
        return self._items[self.top]

    def isEmpty(self) -> bool:
        return self.top == -1
    
    def size(self):
        return self.top +1
    
stack = ArrayStack3(12)
stack.push('a')
stack.push('b')
stack.push('c')
print(stack._items)
print(stack.pop())
print(stack.pop())
print(stack.pop() if stack else None)
```
