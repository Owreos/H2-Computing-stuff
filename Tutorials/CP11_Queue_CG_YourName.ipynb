## Queue ADT

**Queue** is a First in First Out (FIFO) structure in which access is completely restricted to just one end – this end is known as top.

### Operations
The basic operations of a queue areto enqueue and dequeue item from its top. 
* **enqueue**: Add an item at the end of the queue.
* **dequeue**: Remove and return the first item of the queue, if the queue is not empty

Other supporting functions to be added are:
* **isEmpty()**: Returns true if the queue is empty, otherwise false
* **__size()**: Returns the number of items in the queue
* **peek()**: return the first item of the queue without removing it, if the queue is not empty

#### Implement ArrayQueue2 with the following :
* Maintain a variable `front` that points to the item at the front of the queue. Starts at 0 and advances as items dequeued.
* Rear variable points to the last item at position n-1, where n is the number of items in queue
* The items will be shifted to the start of the queue when the rear pointer is about to run off the end

```python
class ArrayQueue2:
    def __init__(self, cap : int):
        self.arr = [None] * cap
        self.capacity = cap
        self.front = 0
        self.rear = -1

    def dequeue(self):
        if  self.is_empty():
            raise Exception("queue is empty")
        popped, self.arr[self.front] = self.arr[self.front], None
        self.front += 1    
        # print(self.arr)
        return popped

    def enqueue(self, item):
        # print(self.rear, self.capacity)
        if self.rear + 1 == self.capacity:
            if self.front == 0:
                raise Exception("Queue Overflow!")
            self.shifthelper(item)
        self.rear += 1
        self.arr[self.rear] = item
        # print(self.arr)

    def shifthelper(self, item):
        for pos, value in enumerate(self.arr[self.front:self.rear + 1]):
            self.arr[pos] = value
        self.arr[self.rear] = None
        self.rear -= self.front
        self.front = 0

    def peek(self):
        return self.arr[self.front]
    
    def is_empty(self):
        return self.rear - self.front == 0 or self.rear == -1 and self.front == 0
```

```python
class Node:
    def __init__(self, data, nxt: object = None):
        self.data = data
        self.next = nxt

    # getters:
    def getData(self):
        return self.data;
    
    def getNext(self):
        return self.next;
    
    #setters:
    def setData(self, nData):
        self.data = nData;

    def setNext(self, nNext):
        self.next = nNext;

    def __str__(self):
        return "data: {} next: {} ".format(self.getData(), self.getNext()) if self.getNext() else "data: {} next: None".format(self.getData())
```

* Code `isEmpty()`, `size()` and `peek()` methods 


```python
class LinkedListQueue:
    def __init__(self):
        self.front = None
        self.rear = None
        self._size = 0  

    def enqueue(self, nnode: object):
        if not self.front:
            self.front = nnode
            self.rear = nnode
        else:
            self.rear.next = nnode
            self.rear = nnode
        self._size += 1

    def dequeue(self):
        if self._size == 0:
            raise Exception("Queue Underflow!")
        popped = self.front.data
        self.front = self.front.next
        self._size -= 1
        return popped
    
    def peek(self):
        return self.front.data

    def size(self):
        return self._size
    
    def is_empty(self):
        return self._size == 0
    
    def __str__(self):
        res = ""
        probe = self.front
        while probe:
            res += probe.data
            res += " "
            probe = probe.next

        return res
```
