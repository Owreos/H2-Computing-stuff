1. Write a recursive function that accepts an integer argument, n. The function should display n lines of asterisks on the screen, with the first line showing 1 asterisk, the second line showing 2 asterisks, up to the nth line which shows n asterisks.

````python
def pyramid(n: int, count = 0):
    if count == n:
        return 
    genrow = ' '.join(['*' for _ in range(count)])
    print('{:^{}}'.format(genrow, n*3))
    pyramid(n, count+1)
pyramid(10)
````

2. Write a recursive function that accepts two integer arguments, low and high. The function should return the sum of the numbers from low up to high. For example, if 2 and 10 are passed as arguments, the function will return the sum of 2, 3, 4, . . . ,10.

````python
def highlowsum(low: int, high: int):
    if low == high:
        return high
    return low + highlowsum(low+1, high)
#testcases
print(highlowsum(2, 5))
print(highlowsum(1, 10))
print(highlowsum(1,3))
````

3.Write a program that asks the user to enter 3 test scores. The program should display a letter grade for each score and the average test score. Write the following functions in 
the program:
1. calc_average -- This function should accept three test scores as arguments and return the average of the scores.
2. determine_grade -- This function should accept a test score as an argument and return a letter grade for the score, based on the following grading scale:


````python
def calcscore(score: int):
    scores = [-1, 59, 69, 79, 89, 100]
    grade = ['F', 'D', 'C', 'B', 'A', ]
    for x in range(0, len(scores)):
        if scores[x-1] < score <= scores[x]:
            return grade[x-1]
    return 'invalid'
#testcases
print(calcscore(-10))
print(calcscore(90))
print(calcscore(0))
print(calcscore(77)) # if this is 77 then...
````

4. Write a menu-driven automatic teller program in which the user’s bank balance is initialized to $1,000.  The user should be allowed to perform as many transactions as he or she wishes from the menu.

```
 1.	Deposit
 2.	Withdrawal
 3.	See balance
 4.	Quit
```
For the withdrawal option, the user should be prompted to select from the choices `$50`, 	`$100`, `$200`, and `$500`.  (Do not allow user to overdraw.)  After each transaction, the 	program should print the current balance.  On exiting, the program should print a 	courteous message. In this program, design a function for each option. 


````python
class Account:
    def __init__(self):
        self._balance = 0

    def deposit(self, n: int) -> None:
        self._balance += n

    def withdraw(self, n: int) -> bool:
        if n > self._balance:
            print("Insufficient funds!")
            return False
        self._balance -= n
        return True

    def check_balance(self) -> int:
        return self._balance


class Bank:
    def __init__(self):
        self.accounts = {}

    def create_account(self, name: str) -> None:
        if name in self.accounts:
            print("Account already exists!")
        else:
            self.accounts[name] = Account()
            print(f"Account '{name}' created successfully!")

    def init_banking(self) -> None:
        while True:
            if not self.accounts:
                ans = input("No accounts found. Create an account? (y/n): ").lower()
                if ans == 'y':
                    name = input("Enter account name: ")
                    self.create_account(name)
                else:
                    print("Exiting... Goodbye!")
                    break

            print('''\nWhat would you like to do?\n
1. Check your balance
2. Deposit
3. Withdraw
4. Create a new account
5. Exit
''')
            try:
                choice = int(input("Enter your choice (1-5): "))
            except ValueError:
                print("Invalid input. Please enter a number between 1 and 5.")
                continue

            if choice == 5:
                print("Thank you for using the banking system. Goodbye!")
                break
            elif choice == 4:
                name = input("Enter new account name: ")
                self.create_account(name)
            else:
                name = input("Enter your account name: ")
                if name not in self.accounts:
                    print("Account not found!")
                    continue

                if choice == 1:
                    print(f"Current balance: {self.accounts[name].check_balance()}")
                elif choice == 2:
                    try:
                        amount = int(input("Enter amount to deposit: "))
                        self.accounts[name].deposit(amount)
                        print("Deposit successful!")
                    except ValueError:
                        print("Invalid input. Please enter a valid number.")
                elif choice == 3:
                    try:
                        amount = int(input("Enter amount to withdraw: "))
                        if self.accounts[name].withdraw(amount):
                            print("Withdrawal successful!")
                    except ValueError:
                        print("Invalid input. Please enter a valid number.")
                else:
                    print("Invalid choice. Please try again.")

# Initialize the banking system
bank = Bank()
bank.init_banking()
````

5. Design a recursive function that accepts two arguments into the parametrs x and y. The function should return the value of x times y. Remember, multiplication can be performed as repeated addition as follows:
 
                  7 x 4 = 4 + 4 + 4 + 4 + 4 + 4 + 4
                  
   (To keep the function simple, assume that x and y will always hold positive integers.)

````python
mult = lambda count, n: n if count == 1 else mult(count-1, n) + n # base case where count = 1 return n else function(count-1, number) + number itself
print(mult(7, 4))
````

6. Design a function that accepts an integer argument and returns the sum of all the integers from 1 up to the number passed as an argument. For example, if 50 is passed as an argument, the function will return the sum of  1, 2 , 3, . . . ,50. Use recursion to calculate the sum. 

````python
def returnsum(m: int, i = 1):
    if i == m:
        return i
    return returnsum(m, i+1) + i
#testcases
print(returnsum(10))
print(returnsum(100))
````

7. Write a function that takes three numbers as parameters, and returns the median value of those parameters as its result. Include a main program that reads three values from the user and displays their median. Do not use any built-in functions. 
Hint: The median value is the middle of the three values when they are sorted into ascending order.

````python
import random
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    piv = arr[0]
    high = [x for x in arr if x > piv]
    low = [x for x in arr if x < piv]
    mid = [x for x in arr if x == piv]
    return quicksort(low) + mid + quicksort(high)

def medianforeverything(arr):
    sorarr = quicksort(arr)
    return sorarr[len(sorarr)//2]

#testcase
print(medianforeverything([10, 8, 9, 10, 2, 3, 3, 3])) #this probably isnt the desired solution tbh i mean it works???
print(medianforeverything([int(input('val1: ')), int(input('val2: ')), int(input('val3: '))]))
````

9.  Write a function that generates a random password. The password should have a random length of between 7 to 10 characters. Each character should be randomly selected from position 33 to 126 in the ASCII table. Your function will not take any parameters. It will return the randomly generated password as its only result. Display the randomly generated password in the main program.

````python
import random
import string
def genpass() -> str:    
    length = random.randint(7, 10)
    return ''.join([str(chr(random.randint(33, 126))) for _ in range(length)])
print(genpass())
````

10. Write a function that determines whether or not a password is good. A good password is a one that is at least 8 characters long and contains at least one uppercase letter, at least one lowercase letter, and at least one digit. Your function should return True if the password passed to it as its only parametere is good. Otherwise it should return False. Include a main program that reads a password from the user and reports whether or not it is a good.

````python
def checksec(password: str) -> bool:
    if len(password) < 8:
        return False
    if password.isupper() or password.islower():
        return False
    return True if [x for x in password if x.isalpha()] else False #Checks if there is actually are letters inside the password
#testcases
print(checksec('1234567'))  # False
print(checksec('12345678')) # False
print(checksec('12345678t'))# False
print(checksec('123456Tt')) # True
````
