###### Task 1

Implement the class Node which has the following attributes:
* `_data`
* `_next`

and methods:
* `getData(self)` returns _data of the Node object.
* `getNext(self)` returns _next of the Node object.
* `setData(self)` set the value of _data of the Node object.
* `setNext(self)`  set the value of _next of the Node object.
* `__str__`  returns the string representation of the Node object.
* `__init__`  accepts two arguments 'data', 'nxt' and initialize the Node object with the value passed in.

Write codes to test the Node class.

```python
class Node:
    def __init__(self, data, nxt: object = None):
        self._data = data
        self._nxt = nxt

    # getters:
    def getData(self):
        return self._data;
    
    def getNext(self):
        return self._nxt;
    
    #setters:
    def setData(self, nData):
        self._data = nData;

    def setNext(self, nNext):
        self._nxt = nNext;

    def __str__(self):
        return "data: {} next: {} ".format(self.getData(), self.getNext()) if self.getNext() else "data: {} next: None".format(self.getData())
```

Implement the class `SinglyLinkedList` which has the following attributes:

and methods:
* `getHead()` returns _head of the SinglyLinkedList object.
* `setHead()` set the value of _head of the SinglyLinkedList object.
* `__str__()` returns the string representation of the SinglyLinkedList object.
* `__init__()` initializes the SinglyLinkedList object to an empty list.
* `search()` accepts an argument: `target`. If `target` is found, returns the position , otherwise returns -1.
* `replace()` accepts two arguments: `old`, `new`. Replace the first node with `old` data value with `new` data value.
* `replaceAll()` accepts two arguments: `old`, `new`. Replace all nodes with `old` data value with `new` data value.
* `replaceByPosition()` accepts two arguments: `pos`, `new`. Replace all nodes with `old` data value with `new` data value.
* `isEmpty()` returns True if list is empty, otherwise, False.
* `size()` returns the number of nodes in the list.
* `insert()` accepts two arguments: `pos`, `node`. Insert `node` at the position indicated by `pos`. Assume the first node is at position zero. If `pos`> length-1, append `node`  
* `remove()` accepts an arguments: `pos`. Remove `node` at the position indicated by `pos`. If `pos`> length-1, remove the last `node` 
* `append()` accepts an arguments: `node`. Append `node` at the end of the list.

Write codes to test the SinglyLinkedList class.

```python
class SinglyLinkedList:
    def __init__(self):
        self._head = None

    def getHead(self):
        return self._head

    def setHead(self, head):
        self._head = head

    def isEmpty(self):
        return self._head is None

    def size(self):
        probe = self._head
        count = 0
        while probe:
            count += 1
            probe = probe.getNext()
        return count

    def search(self, target):
        probe = self._head
        pos = 0
        while probe:
            if probe.getData() == target:
                return pos
            pos += 1
            probe = probe.getNext()
        return -1

    def replace(self, old, new):
        probe = self._head
        while probe:
            if probe.getData() == old:
                probe.setData(new)
                return
            probe = probe.getNext()
        return -1

    def replaceAll(self, old, new):
        probe = self._head
        while probe:
            if probe.getData() == old:
                probe.setData(new)
            probe = probe.getNext()

    def replaceByPosition(self, pos: int, new):
        probe = self._head
        count = 0
        while probe:
            if count == pos:
                probe.setData(new)
                return
            count += 1
            probe = probe.getNext()
        return -1  

    def insert(self, pos, node: object):
        if pos == 0:
            node.setNext(self.getHead())
            self.setHead(node)
            return
        probe = self._head
        for i in range(pos - 1):
            if not probe:
                return -1  
            probe = probe.getNext()
        node.setNext(probe.getNext())
        probe.setNext(node)

    def remove(self, pos: int):
        if not self._head:
            return -1  

        if pos == 0:
            self._head = self._head.getNext()
            return

        if pos >= self.size():
            return -1  

        pointer = self._head
        for i in range(pos - 1):
            pointer = pointer.getNext()
        pointer.setNext(pointer.getNext().getNext())

    def append(self, node: object):
        if not self._head:
            self._head = node
            return
        probe = self._head
        print(probe)
        while probe.getNext() is not None:
            probe = probe.getNext()
        probe.setNext(node)

    def __str__(self):
        return self._strHelper(self._head)

    def _strHelper(self, node):
        result = ""
        while node:
            result += str(node.getData()) + " -> "
            node = node.getNext()
        return result + "None"
```
