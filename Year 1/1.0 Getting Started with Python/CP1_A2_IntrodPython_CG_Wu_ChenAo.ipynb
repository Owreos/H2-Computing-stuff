1. Write a program that accepts the user’s name (as text) and age (as an integer) as input. The 
program should output a sentence containing the user’s name and current age, and age in 
10 years’ time.

```python
def getage():
    name = str(input('Your name: '))
    age = int(input('Your age'))
    return name, age

def years10():
    name, age = getage()
    print(f'{name} of {age} years old you will be {age+10} in ten years time')
years10()
```

3. An employee’s total weekly pay equals the hourly wage multiplied by the total number of 
regular hours plus overtime pay. Overtime pay equals the overtime hours multiplied by 1.5 
times the hourly wage. Write a program that takes as inputs the hourly wage, total regular 
hours, and total overtime hours and display an employee’s total weekly pay (round to the 
nearest dollar). 

```python
def paycalc():
    pay, hours, overhours = int(input('pay: ')), int(input('total hours: ')), int(input('total overtime hours: '))
    print(f'{pay*hours + overhours*1.5*pay:.0f}')

paycalc()
```

9. Compute the approximate age of an individual in seconds within 99% accuracy of results. Your program must work for dates of birth from January 1, 1900 to present. Provide comments on your codes and test cases to make sure that the approximate age in seconds is withn 99% accuracy. Note: You are only allow to use datetime module to obtain the current date. 


```python
import datetime
def get_time(byear, bmonth, bday, bseconds):
    currdate = datetime.datetime.now()
    cyear, cmonth, cday, chour, cmin, csec = currdate.year, currdate.month, currdate.day, currdate.hour, currdate.minute, currdate.second # gets current year, month and day from the datetime.date.today() object
    cseconds = chour*60*60 + cmin*60 + csec
    tyear, tmonth, tday= cyear - byear - 1, 12 - bmonth + cmonth -1, 30 - bday + cday,
    tsec = 86400 - bseconds + cseconds # logic is years passed is cyear - byear, remaining bmonths from next year + cmonth, and remaining bday from next month + cday
    if tsec >= 86400:
        tsec -= 86400
        tday += 1
    if tday >= 30:# checks if tday is more than 30, if so add one more month to tmonth
        tday -= 30# remove excess days
        tmonth += 1
    if tmonth >= 12:# checks if tmonth is more than 12, if so, add one more to tyear
        tmonth -= 12# remove excess days
        tyear += 1
    return tyear, tmonth, tday, tsec

def get_totaltime():
    year = int(input('what year were you born in?: '))
    month = int(input('what month were you born in?: '))
    day = int(input('what day were you born on?: '))
    time = (input('what time were you born on?(format hh:mm:ss)): ')).split(':')
    sec = int(time[0])*60*60 + int(time[1])*60 + int(time[2])
    res = get_time(year, month, day, sec)
    print('It has been about  {} years {} months and {} days and {}s since you were born, how long!'.format(res[0], res[1], res[2], res[3]))
get_totaltime()
```
