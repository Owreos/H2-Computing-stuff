1. In a two-person game of Rock-Scissors-Paper, each player selects either ‘R’, ‘S’ or ‘P’. The winner is determined as follows: Rock breaks Scissors. Paper covers Rock. Scissors cuts Paper. The game is a tie if both players select the same choice. Write a program to allow players make their input and output the winner.

````python
def rps(p1: str, p2: str) -> int:
    plyinp = [p1.lower(), p2.lower()]
    if p1.lower() == p2.lower():
        return 0
    if 'r' in plyinp and 's' in plyinp:
        return plyinp.index('r') + 1
    if 'r' in plyinp and 'p' in plyinp:
        return plyinp.index('p') + 1
    if 'p' in plyinp and 's' in plyinp:
        return plyinp.index('s') +1
print(rps('s', 'p'))
print(rps('s','r'))
````

2. Write a loop that that prints the first 128 ASCII values followed by the corresponding characters.

````python
def asci(m):
    aslis = [f'{x}: {chr(x)}' for x in range(m+1)] #create a list of index and their corresponding ascii 
    return '\n'.join(aslis) # format it correctly
print(asci(128))
````

4. Write a loop that counts the number of space characters in a string. Recall that the space character is represented as " ".

````python
def countspace(x: str) -> int:
    return len([_ for _ in x if _ == ' '])
#testcases:
print(countspace('nd uaksd ioasdioha siudiu'))          #3
print(countspace('1 2 3 4'))                            #3
print(countspace('1ksadb sda knsk sjn jsn djnsdj '))    #6
print(countspace('The red fox jumps over the lazy dog'))#7
````

5. The German mathematician Gottfried Leihniz developed the following method to approximate the value of 
```
    π/4 = 1 – 1/3 + 1/5 – 1/7 + . . . . . .
```
Write a program that allows the user to specify the number of iterations used in this 
approximation and that displays the resulting value.

````python
def approxpi(cycles: int):
    leiseries = [((-1) ** x) / (x * 2 + 1) for x in range(cycles) ] 
    #okok ill explain this creates a series of length x of odd numbers based on 2x + 1
    #and series is creates such that the sign alternates based on (-1)**x where it is negative when x = 2n + 1 and positive when 2x
    return sum(leiseries)*4

print(approxpi(10000))
````

7. Write a program to print a multiplication table based on the number entered. A sample run (with input 5) is shown below:
    
```
Number of Columns: 5
        1  2  3  4  5
    1   1
    2   2  4
    3   3  6  9
    4   4  8  12 16
    5   5  10 15 20 25

```

````python
def multtable(size):
    #print a header
    print(f'número de columnas: {size}')
    length = len(str(size**2))
    header = [' ' * length] + ['{:<{}}'.format(x, length) for x in range(1, size)]
    print(' '.join(header))

    for y in range(1, size): # iterate based on rows
        rowlabel =  ['{:<{}}'.format(y, length)] # pad create row label
        row = ['{:<{}}'.format(x*y, length) for x in range(1, y+1)] # create a series for the row and pad on the right until length 3
        print(' '.join(rowlabel + row))
multtable(100)
````

9. The greatest common divisor of two positive integers, A and B, is the largest number that 	can be evenly divided into both of them. Euclid’s algorithm can be used to find the 	greatest common divisor (GCD) of two positive integers. You can use this algorithm in 	the following manner:

```
(a)	Compute the remainder of dividing the larger number by the smaller number.

(b)	Replace the larger number with the smaller number and the smaller number with the
    remainder.

(c)	Repeat this process until the smaller number is zero.

The larger number at this point is the GCD of A and B. Write a program that lets the user
enter two integers and then prints each step in the process of using the Euclidean algorithm
to find their GCD.

```

````python
def GCD(n1: int, n2: int) -> int:
    n1 = sorted([n1, n2])[1]
    n2 = sorted([n1, n2])[0]
    while n2:
        n1, n2 = n2, n1 % n2
    return n1

print(GCD(30, 15))  # Output: 2
````
