1. A class has at most 40 students. Write a program that will read in the 
scores of an exam for each student in the class, finds and prints the class average and 
then prints each of the scores that are above the class average. Use an array to process the inputs from the user.
For example, if the input scores are as follow:
```
    70
    80
    71
    58
    79
    92
```
the output would be
```
    class average: 75.0
    scores above average: 80 79 9
 
```

````python
def classav(max_stud):
    scores = [0]*max_stud
    i = 0
    score = ''
    while i < max_stud and score != -1:
        score = int(input('Enter score(-1 to terminate): '))
        scores[i] = score
        i += 1
    if score == -1:
        scores = scores[0:i-1]
        i -= 1
    print(scores)
    av = sum(scores)/i
    print(f'You have input {i} students')
    print(f'Class average: {av:.1f}')
    print('scores above average: ', end='')
    for y in scores:
        if y > av:
            print(y, end = ' ')
        
classav(10)
````

3. Write a program to read the close price in the data files of several stocks for each day into two parallel arrays. Use the data for the Mar 2023. For each stock, the program should print the maximum and minimum price and the day it was achieved. 


````python
import csv
def findminmax(filepath: str, yearmonth: str):
    with open(filepath, 'r') as f:
        read = csv.reader(f, delimiter = ',')
        header = next(read)
        data = [r for r in read if yearmonth in r[0]]
        print(data)
    # prices = [row[]]
    dates = [row[0] for row in data]
    prices = [float(row[4]) for row in data]
    minp, maxp = min(prices), max(prices)
    mindate, maxdate = dates[prices.index(minp)], dates[prices.index(maxp)]
    return (minp, mindate), (maxp, maxdate)

res = findminmax('C52.csv', '2023-03')
print('min:', res[0][0], 'mindate:', res[0][1], 'max:', res[1][0],'maxdate:', res[1][1])
````

4. Given an array of integers, return a new array such that each element at index i of the new array is the product of all the numbers in the original array except the one at i.

````python
from functools import reduce
def sumexcepti(arr: list):
    return [reduce(lambda x, y: x*y, arr[:x] + arr[x+1:]) for x in range(len(arr))]

print(sumexcepti([1, 2, 3, 4, 5]))

:3
````
