```python
class ATM:
    def __init__(self):
        self._balance = 0

    def deposit(self, n: int) -> None:
        self._balance += n

    def withdraw(self, n: int) -> bool:
        if n > self._balance:
            print("Insufficient funds!")
            return False
        self._balance -= n
        return True

    def check_balance(self) -> int:
        return self._balance


class Bank:
    def __init__(self):
        self.accounts = {}

    def create_account(self, name: str) -> None:
        if name in self.accounts:
            print("Account already exists!")
        else:
            self.accounts[name] = ATM()
            print(f"Account '{name}' created successfully!")

    def init_banking(self) -> None:
        while True:
            if not self.accounts:
                ans = input("No accounts found. Create an account? (y/n): ").lower()
                if ans == 'y':
                    name = input("Enter account name: ")
                    self.create_account(name)
                else:
                    print("Exiting... Goodbye!")
                    break

            print('''\nWhat would you like to do?\n
1. Check your balance
2. Deposit
3. Withdraw
4. Create a new account
5. Exit
''')
            try:
                choice = int(input("Enter your choice (1-5): "))
            except ValueError:
                print("Invalid input. Please enter a number between 1 and 5.")
                continue

            if choice == 5:
                print("Thank you for using the banking system. Goodbye!")
                break
            elif choice == 4:
                name = input("Enter new account name: ")
                self.create_account(name)
            else:
                name = input("Enter your account name: ")
                if name not in self.accounts:
                    print("Account not found!")
                    continue

                if choice == 1:
                    print(f"Current balance: {self.accounts[name].check_balance()}")
                elif choice == 2:
                    try:
                        amount = int(input("Enter amount to deposit: "))
                        self.accounts[name].deposit(amount)
                        print("Deposit successful!")
                    except ValueError:
                        print("Invalid input. Please enter a valid number.")
                elif choice == 3:
                    try:
                        amount = int(input("Enter amount to withdraw: "))
                        if self.accounts[name].withdraw(amount):
                            print("Withdrawal successful!")
                    except ValueError:
                        print("Invalid input. Please enter a valid number.")
                else:
                    print("Invalid choice. Please try again.")

# Initialize the banking system
bank = Bank()
bank.init_banking()
```

    If it’s all lowercase - converts it to all uppercase (”hello” → “HELLO”)
    If it’s all uppercase - converts it to all lowercase (”HELLO” → “hello”)
    Only if there’s an “e” in the string - prints the number of times “e” appears in the string. (Regardless of upper or lower case letters)
    Only if there’s the word “python” in the string - changes it to “python!” (only strictly for "python" and "PYTHON", "Python" does not count)
    Only it’s a number - multiplies it by 5 (”5” → 25)
    It must be able to handle cases when there is a mix of upper and lower case characte

```python
def prosstr(x: str)->str:
    if x.isupper():
        x.lower()
    elif x.islower():
        x.upper()
    if 'e' in x:
        print(f'e appears {x.count('e')} times')
    if 'python' in x:
        x.replace('python', 'python!')
    if x.isnumeric():
        return int(x) * 5
    return x
```

```python
def checkip(ip):
    arr = ip.split(' ')

    if len(arr) == 4:
        for i in arr:
            if i.isnumeric() and 0 <= int(i) <= 255 and str(int(i)) == i:
                continue
            else:
                return False
    else: return False

    return True

```

```python
# Valid IP addresses
print(checkip("192 168 1 1"))  # Expected: True (valid IP)
print(checkip("255 0 255 255"))  # Expected: True (valid IP)

# Invalid IP addresses
print(checkip("256 168 1 1"))  # Expected: False (one segment exceeds 255)
print(checkip("192 168 01 1"))  # Expected: False (leading zero not allowed)
print(checkip("192 168 1"))  # Expected: False (not enough segments)
print(checkip("192 168 1 a"))  # Expected: False (non-numeric segment)

# Edge cases
print(checkip(""))  # Expected: False (empty input)
print(checkip("0 0 0 0"))  # Expected: True (all segments are zero)
print(checkip("255 255 255 255"))  # Expected: True (upper limit valid IP)
print(checkip("300 300 300 300"))  # Expected: False (all segments exceed 255)
```

```python
movies = ["The Shawshank Redemption", "The Godfather", "The Dark Knight", "Forrest Gump", "Inception", "The Matrix", "Avengers: Infinity War", "Back to the Future", "The Lion King", "Pulp Fiction"]

def searchmovies(movies):
    # Take input and split it into search terms
    terms = input("Enter search terms: ").lower().split(' ')
    # Search for movies containing all terms (case-insensitive)
    res = [movie for movie in movies if all(term in movie.lower() for term in terms)]
    # Print results
    print(res, len(res))

# Call the function
searchmovies(movies)
```
