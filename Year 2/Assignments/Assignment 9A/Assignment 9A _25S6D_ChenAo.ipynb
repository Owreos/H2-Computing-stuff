1. We want to use linear search to count the number of occurrences of a key in a list. When 
the key does not exist in the list, it is 0 occurrence. Write a recursive function 
RecSearch(A,key,start) which requires three parameters, the list A, the key to be 
found, and the start index to search in the list. This function returns the number of 
occurrences of this key in the list.


````python
def RecSearch(A, key, start):
    if start >= len(A):
        return 0
    count = 1 if A[start] == key else 0 
    return count + RecSearch(A, key, start + 1)
````

````python

# Test cases for RecSearch function
test_cases = [
    ([1, 2, 3, 2, 4, 2], 2, 0, 3),  # key 2 appears 3 times
    ([5, 5, 5, 5], 5, 0, 4),  # all elements match
    ([1, 2, 3, 4, 5], 6, 0, 0),  # key not found
    ([], 1, 0, 0),  # empty list
    ([10], 10, 0, 1),  # single element match
    ([10], 20, 0, 0),  # single element no match
]

# Run test cases
for A, key, start, expected in test_cases:
    result = RecSearch(A, key, start)
    status = "✓" if result == expected else "✗"
    print(f"{status} RecSearch({A}, {key}, {start}) = {result} (expected {expected})")
````

2. Write a recursive function BinSearch(A,low,high,key) which searches a sorted 
list A, A[low] . . . . . A[high] , for a match with key using binary search. Return the 
index of the matching item or −1 if the key is not found.

````python
def BinSearch(A, key, low, high):
    if low > high:
        return -1

    mid = (low + high) // 2
    if mid ==key:
        return mid
    if key < mid:
        return BinSearch(A, key, low, mid+1)
    if key > mid:
        return BinSearch(A, key, mid-1, high)
````
