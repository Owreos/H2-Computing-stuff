#### Task 1

Write a function: **partition(A, low, high)** that accepts 3 arguments, array A, low index and high index, and returns the pivot position of the array.

* take the middle element of the array as a pivot
* find the position of the pivot:
    * swap the pivot value from the middle index to the low index of the array.
    * assign a left and right index to next index after the low index and high index respectively
    * move the left index to the right until a value greater than the pivot value is found
    * move the right index to the left until a value less than the pivot value is found
    * if left index is less than the right index, swap the their values
    * otherwise, swap the value of the right index with the value of the pivot 
    * return pivot index


Test your function with the following values:
```
A = [13, 12, 57, 99, 39, 64, 59, 48, 70]
low = 0
high = 8
```

````python
def partition(A, low, high):
    mid = (low + high) // 2
    pivot = A[mid]
    A[low], A[mid] = A[mid], A[low]
    left = low + 1
    right = high
    while True:
        while A[left] <= pivot and left <= right:
            left += 1
        while A[right] >= pivot and left <= right:
            right -= 1
        if A[left] > A[right]:
            A[left], A[right] = A[right], A[left]
        else:
            break
    A[low], A[right] = A[right], A[low]
    return right

A = [13, 12, 57, 99, 39, 64, 59, 48, 70]
low = 0
high = 8
print(partition(A, low, high))
            
````

Implement class WordIntPair with the following attributes and methods.

    *  word, number
 
 
* ``compareWord(self, obj)`` method accepts another WordIntPair object and returns the follwoing:

    * -1 if its word is less than 


* ``compareNum(self, obj)`` method accepts another WordIntPair object and returns the follwoing:
    * 1 if its number is greater than
    * 0 if its number is equal
    * -1 if its number is less than 
    

````python
A = [31, 26, 54, 17, 44, 20, 77, 93, 55]
low = 0
high = 8

class WordIntPair:
    def __init__(self, word, number):
        self.word = word
        self.number = number
    
    def compare_word(self, obj):
        if self.word == obj.word:
            return 0
        elif self.word > obj.word:
            return 1
        else:
            return -1

    def compare_num(self, obj):
        if self.number == obj.number:
            return 0
        elif self.number > obj.number:
            return 1
        else:
            return -1
````

#### Task 5.5
Write a function ```task5_5(A, sortBy,  asc)``` that accepts 3 arguments.

The function sorts the array of WordIntPair ```A``` using **insertion sort algorithm**. 

The argument ```sortBy``` allows 2 values, 'word' or 'number' which indicates which attribute 
the WordIntPir should be sorted.  

If ```asc``` is True, the function will sort array A in ascending order ,otherwise descending order. 

The argument ```sortBy``` has a default value ```'work'``` and  ```asc``` has default value of True.

````python

def task5_5(A, sortBy = "word", asc = True):
    compare = (lambda a, b: a.compare_word(b)) if sortBy == "word" else (lambda a, b: a.compare_num(b))
    for i in range(1, len(A)):
        key = A[i]
        j = i-1
        while j >= 0 and ((asc and compare(A[j], key) == 1) or (not asc and compare(A[j], key) == -1)):
            A[j+1] = A[j]
            j-=1
        A[j+1] = key
    
    return A
````

#### Task 5.7

Write a function task5_7(A, sortBy,  asc) that accepts 3 arguments.

The function sorts the array of WordIntPair A using ***bubble sort algorithm***. 

The argument sortBy allows 2 values, 'word' or 'number' which represents which attribute the WordIntPir should be sorted.

If ```asc``` is True, the function will sort array A in ascending order, otherwise descending order. 

The argument ```sortBy``` has default value ```'work'``` and ```asc``` has default value of True.

````python
def task5_7(A, sortBy="word", asc=True):
    compare = (lambda a, b: a.compare_word(b)) if sortBy == "word" else (lambda a, b: a.compare_num(b))

    last_exchange = len(A) - 1

    while last_exchange > 0:
        c = 0
        for i in range(1, last_exchange + 1):
            cmp = compare(A[i - 1], A[i])
            if (asc and cmp == 1) or (not asc and cmp == -1):
                A[i - 1], A[i] = A[i], A[i - 1]
                c = i

        last_exchange = c

    return A
````

#### Task 5.8

Write a function task5_8(A, low, high, sortBy, asc) that accepts 5 arguments.

The function sorts the array of WordIntPair A using ***merge sort algorithm***. 

The argument ```sortBy``` allows 2 values, 'word' or 'number' which represents which attribute the WordIntPir should be sorted. 

If ```asc``` is True, the function will sort array A in ascending order, otherwise descending order. 

The argument ```sortBy``` has default value ```'work'``` and ```asc``` has default value of True.

````python
def merge(A, l, m, r, compare, asc):
    s = r - l + 1
    temp = [None] * s

    left = l
    right = m + 1
    i = 0

    while left <= m and right <= r:
        cmp = compare(A[left], A[right])

        if (asc and cmp != 1) or (not asc and cmp != -1):
            temp[i] = A[left]
            left += 1
        else:
            temp[i] = A[right]
            right += 1

        i += 1

    while left <= m:
        temp[i] = A[left]
        left += 1
        i += 1

    while right <= r:
        temp[i] = A[right]
        right += 1
        i += 1

    for i in range(s):
        A[l + i] = temp[i]
    
def task5_8(A, low, high, sortBy = "word", asc = True):
    compare = (lambda a, b: a.compare_word(b)) if sortBy == "word" else (lambda a, b: a.compare_num(b))

    def merge_sort(A, l, r):
        if l < r:
            m = (l + r) // 2
            merge_sort(A, l, m)
            merge_sort(A, m + 1, r)
            merge(A, l, m, r, compare, asc)

    merge_sort(A, low, high)
    return A
````
