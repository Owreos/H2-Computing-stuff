Write a function **bubbleSort(A)** that accepts 1 argument.

* A - an array

and returns a sorted array.

It sorts the array in ascending order using bubble sort algorithm (without using the last exchange index) and stop when the array is sorted.

Use a **for loop** or **while loop** to implement the sorting algorithm.

Write test case to test your function using the array: 

```
[38, 58, 13, 15, 51, 27, 10, 19, 12, 86, 49, 67, 84, 60, 25]
```

````python
A = [38, 58, 13, 15, 51, 27, 10, 19, 12, 86, 49, 67, 84, 60, 25]
def bubbleSort(A):
     n = len(A)
     for i in range(n):
          swapped = False
          for z in range(n - i - 1):
               if A[z] > A[z + 1]:
                    A[z], A[z + 1] = A[z + 1], A[z]
                    swapped = True
          if swapped == False:
               return A
          
print(bubbleSort(A))
````

#### Task 2

Write a function **bubbleSort2(A)** that accepts 1 arguments.

* A - an array

and returns a sorted array.

It sorts the array in ascending order using bubble sort algorithm and stop when the last exchange index is 0 after a pass. 

Write test case to test your function using the array: 
```
[38, 58, 13, 15, 51, 27, 10, 19, 12, 86, 49, 67, 84, 60, 25]
```


````python
A = [38, 58, 13, 15, 51, 27, 10, 19, 12, 86, 49, 67, 84, 60, 25]
def bubbleSort2(A):
    n = len(A)
    while n > 1:
        last_exchange_index = 0
        for i in range(1, n):
            if A[i - 1] > A[i]:
                A[i - 1], A[i] = A[i], A[i - 1]
                last_exchange_index = i
        n = last_exchange_index
        if n == 0:
            break
    return A
print(bubbleSort2(A))
````

#### Task 3

Write a function **insertionSort(A)** that accepts 1 arguments.

A - an array

It sorts the array in ascending order using insertion sort algorithm and returns a sorted array. 

Write test case to test your function using the array: [38, 58, 13, 15, 51, 27, 10, 19, 12, 86, 49, 67, 84, 60, 25]


````python
A = [38, 58, 13, 15, 51, 27, 10, 19, 12, 86, 49, 67, 84, 60, 25]
def insertionSort(A):
    for i in range(1, len(A)):
        key = A[i]
        j = i - 1
        while j >= 0 and A[j] > key:
            A[j + 1] = A[j]
            j -= 1
        A[j + 1] = key # basically only insert when theres a space hence the name :3c
    return A

print(insertionSort(A))
````

##### Task 4.3
Write a function ```task4_3(A, asc)``` that accepts 2 arguments. 

The function sorts the array ```A``` using **insertion sort algorithm**. 

If ```asc``` is True, the function will sort array A in ascending order ,otherwise descending order. 

The argument ```asc``` has a default value of True.



````python
A = [38, 58, 13, 15, 51, 27, 10, 19, 12, 86, 49, 67, 84, 60, 25]
def task4_3(A, asc: bool):
    
    for i in range(1, len(A)):
        key = A[i]
        j = i - 1
        if asc:
            while j >= 0 and A[j] > key:
                A[j + 1] = A[j]
                j -= 1
        else:
            while j >= 0 and A[j] < key:
                A[j + 1] = A[j]
                j -= 1 

        A[j + 1] = key 
    return A

print(task4_3(A, True))
print(task4_3(A, False))
````

The function sorts the array ```A``` using **bubble sort algorithm**. 

If ```asc``` is True, the function will sort array A in ascending order ,otherwise descending order. 

The argument ```asc``` has a default value of True.
   
 


````python
A = [38, 58, 13, 15, 51, 27, 10, 19, 12, 86, 49, 67, 84, 60, 25]
def task4_4(A, asc: bool):
     n = len(A)
     for i in range(n):
          swapped = False
          for z in range(n - i - 1):
                if asc:
                    if A[z] > A[z + 1]:
                        A[z], A[z + 1] = A[z + 1], A[z]
                        swapped = True
                else:
                    if A[z] < A[z + 1]:
                        A[z], A[z + 1] = A[z + 1], A[z]
                        swapped = True

          if swapped == False:
               return A
print(task4_4(A, True))
print(task4_4(A, False))
````

##### Task 4.7
Write a function ```task4_7(A, key)``` that accepts 2 arguments. 

The function searches key in the array A using **binary search algorithm**.

The function returns the number of comparisons it needs to find key and the index of the ```key``` in the array or -1 if not found. 

````python
def task4_7(A, key):
    low = 0
    high = len(A) - 1
    comparisons = 0
    while low <= high:
        mid = (low + high) // 2
        comparisons += 1
        if A[mid] == key:
            return mid, comparisons
        if key < A[mid]:
            high = mid - 1
        else:
            low = mid + 1

    return -1
````
