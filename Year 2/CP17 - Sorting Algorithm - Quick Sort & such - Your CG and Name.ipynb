Write a function: **partition(A, low, high)** that accepts 3 arguments, array A, low index and high index, and returns the pivot position of the array.

* take the middle element of the array as a pivot
* find the position of the pivot:
    * swap the pivot value from the middle index to the low index of the array.
    * assign a left and right index to next index after the low index and high index respectively
    * move the left index to the right until a value greater than the pivot value is found
    * move the right index to the left until a value less than the pivot value is found
    * if left index is less than the right index, swap the their values
    * otherwise, swap the value of the right index with the value of the pivot 
    * return pivot index


Test your function with the following values:
```
A = [13, 12, 57, 99, 39, 64, 59, 48, 70]
low = 0
high = 8
```
