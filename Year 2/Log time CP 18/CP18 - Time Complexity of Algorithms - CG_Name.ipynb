#### Task 3
Write a function ```insertionSort(A)``` that accepts a list of strings and sort the list of strings in ascending order using **insertion sort algorithm**. 


````python
def insertionSort(A):
    for i in range(len(A)):
        j = i-1 
        key = A[i]

        while j >= 0 and A[j] > key: # important to note for future qq, use key for all comparisons but otherwise full sync + :3
            A[j+1] = A[j]
            j -= 1
        
        A[j + 1] = key
    
    return A 
    
````

````python
# test insertionSort       
A=['x','a','t','z','w','j','m','e','v','q']
insertionSort(A)
for n in A:
   print(n)
````

#### Task 4
Write a function ```quickSort(A)``` that accepts a list of strings and sort the list of strings in ascending order using **quick sort algorithm**. 


````python
# simple quicksort algo

def quickSort(A):
    if len(A) <= 1:
        return A
    high = len(A)
    low = 0
    piv = A[(high + low) // 2]
    left = [x for x in A if x < piv]
    mid = [x for x in A if x == piv]
    right = [x for x in A if x > piv]
    return quickSort(left) + mid + quickSort(right)


# actual weird horse partition in place sorting algo write tomorrow

# def quickSort():
    # pass
````

#### Task 5
Write a function ```bubbleSort(A)``` that accepts a list of strings and sort the list of strings in ascending order using ***bubble sort algorithm***.


````python
def bubbleSort(A):
    c = len(A)
    while c>0:
        for i in range(1, c):
            if A[i-1] > A[i]:
                A[i-1], A[i] = A[i], A[i-1]
        c -= 1
    return A
````

````python
### test bubbleSort
A=['x','a','t','z','w','j','m','e','v','q']
bubbleSort(A)
for n in A:
   print(n)
````

#### Task 6
Write a function ```mergeSort(A)``` that accepts a list of strings and sort the list of strings in ascending order using ***merge sort algorithm***.


````python


def merge(A, l, m, r):
    s = r - l + 1 # size is definitely + 1 since l and r are inclusive
    temp = [None] * s
    left = l
    right = m + 1
    i = 0
    while left <= m and right <= r:
        if A[left] <= A[right]:
            temp[i] = A[left]
            left += 1   
        else:
            temp[i] = A[right]
            right += 1
        i += 1
        
    while left <= m:
        temp[i] = A[left]
        left += 1
        i += 1

    while right <= r:
        temp[i] = A[right]
        right += 1
        i += 1

    for i in range(s):
        A[i + l] = temp[i]
    
def mergeSort(left, right, A):
    if left < right:
        mid = (left + right) // 2
        mergeSort(left, mid, A)
        mergeSort(mid + 1, right, A)
        merge(A, left, mid, right)
        
````

Using the ```timeit``` module and the generated files to compare and contrast the time complexity of the sorting algorithms.

Measure the time it takes for each sorting algorithm to sort the files that contains 1000, 2000 ... to 9000, 10000 strings. Take an average time of 10 runs for each algorithm and file.

````python
# no need to plot graph if you don't have matplotlib module

# import the built-in library timeit 
import timeit

# use the function from Task 2 to read in 1000 strings from the file task1_1000.txt 
t1000 = task2('task1_1000.txt')

# use the timeit function to call the quickSort(t1000) function with the list of 1000 strings, run this just once 

funcs = [
    ("mergeSort", lambda: mergeSort(0, len(t1000)-1, t1000.copy())),
    ("bubbleSort", lambda: bubbleSort(t1000.copy())),
    ("insertionSort", lambda: insertionSort(t1000.copy())),
    ("quickSort", lambda: quickSort(t1000.copy()))
]

ten_tries = [
    [(name, timeit.timeit(func, number=1)) for name, func in funcs]
    for _ in range(10)
]

grouped = list(zip(*ten_tries))

averages = [
    (runs[0][0], sum(x[1] for x in runs) / len(runs))
    for runs in grouped
]

averages.sort(key = lambda x: x[1])
print(averages)
````
