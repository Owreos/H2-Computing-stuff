```python
# chat client example

# set up socket
import socket
chat_socket = socket.socket()
address = input('Enter IPv4 address of server: ')
port = int(input('Enter port number of server: '))
chat_socket.connect((address, port))

while True:
    # receive message
    print('WAITING FOR SERVER...')
    data = b''
    while b'\n' not in data:
        data += chat_socket.recv(1024)
    print('SERVER WROTE: ' + data.decode())
    
    # send message
    data = input('INPUT CLIENT: ').encode()
    chat_socket.sendall(data + b'\n')
```

```python
# Qn 8: use break to exit the program with quit msg

# set up socket
import socket
chat_socket = socket.socket()
address = input("Enter IPv4 address of server: ")
port = int(input("Enter port number of server: "))
chat_socket.connect((address, port))

while True:
    # receive message
    print("WAITING FOR SERVER...")
    data = b''
    while b'\n' not in data:
        data += chat_socket.recv(2014)
    print("SERVER WROTE: " + data.decode())
    
    # exit upon quit msg by server
    if data == b'quit\n':
        break
    
    # send message
    data = input("INPUT CLIENT: ").encode()
    chat_socket.sendall(data + b'\n')
    
    # exit upon quit msg by client
    if data == b'quit':
        break

# close socket
chat_socket.close()
```

```python
# Qn 8: use while loop and Boolean indicator to exit the program with quit msg

# set up socket
import socket
chat_socket = socket.socket()
address = input("Enter IPv4 address of server: ")
port = int(input("Enter port number of server: "))
chat_socket.connect((address, port))

leave = False

while not leave:
    # receive message
    print("WAITING FOR SERVER...")
    data = b''
    while b'\n' not in data:
        data += chat_socket.recv(2014)
    print("SERVER WROTE: " + data.decode())

    if data.decode() == 'quit\n': # exit upon quit msg by server
        leave = True
    else:   
        # send message
        msg = input("INPUT CLIENT: ")
        chat_socket.sendall(msg.encode() + b'\n')
        
        if msg == 'quit': # exit upon quit msg by client
            leave = True

# close socket
chat_socket.close()
```
