```python
# chat server example

# set up socket
import socket
listen_socket = socket.socket()
listen_socket.bind(('127.0.0.1', 6789))
listen_socket.listen()
chat_socket, addr = listen_socket.accept()

while True:
    # send message
    data = input('INPUT SERVER: ').encode()
    chat_socket.sendall(data + b'\n')
    print('WAITING FOR CLIENT...')
    
    # receive message
    data = b''
    while b'\n' not in data:
        data += chat_socket.recv(1024)
    print('CLIENT WROTE: ' + data.decode())
```

```python
# Qn 8: use break to exit the program with quit msg

# set up socket
import socket
listen_socket = socket.socket()
listen_socket.bind(('127.0.0.1',12345))
listen_socket.listen()
chat_socket, addr = listen_socket.accept()

while True:
    # send message
    data = input("INPUT SERVER: ").encode()
    chat_socket.sendall(data + b'\n')
    
    # exit upon quit msg by server
    if data == b'quit':
        break
    
    # receive message
    print("WAITING FOR CLIENT...")
    data = b''
    while b'\n' not in data:
        data += chat_socket.recv(1024)
    print("CLIENT WROTE: " + data.decode())
    
    # exit upon quit msg by client
    if data == b'quit\n':
        break

# close sockets
listen_socket.close()
chat_socket.close()
```

```python
# Qn 8: use while loop and Boolean indicator to exit the program with quit msg

# set up socket
import socket
listen_socket = socket.socket()
listen_socket.bind(('127.0.0.1',12345))
listen_socket.listen()
chat_socket, addr = listen_socket.accept()

leave = False

while not leave:
    # send message
    msg = input("INPUT SERVER: ")
    chat_socket.sendall(msg.encode() + b'\n')

    if msg == 'quit': # exit upon quit msg by server
        leave = True
    else:
        # receive message
        print("WAITING FOR CLIENT...")
        data = b''
        while b'\n' not in data:
            data += chat_socket.recv(1024)
        print("CLIENT WROTE: " + data.decode())
        
        if data.decode() == 'quit\n': # exit upon quit msg by client
            leave = True            

# close sockets
chat_socket.close()
listen_socket.close()
```
