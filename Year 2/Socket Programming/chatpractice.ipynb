The following is a client program for an asymmetric guess-the-number game
where a server generates a random number from 1 to 100 and a client tries to
guess  it within  5  tries.  After each  incorrect guess,  the  server returns  whether 
the guess is greater than or less than the required number:

```python
import socket
chat_socket = socket.socket()
address = input("address: ")
port = int(input("port: "))
chat_socket.connect((address, port))

while True:
    data = b""
    while b"\n" not in data:
        data += chat_socket.recv(1024)
    rec = data.split(b"\n")[0]
    if rec == b"high":
        print("too high!")
    elif rec == b"low":
        print("too low!")
    elif rec == b"guess":
        guess = int(input('Enter guess (1-100): '))
        chat_socket.sendall(str(guess).encode() + b'\n')
    elif rec == b"over":
        print("you lose!")
        break
    elif rec == b"win":
        print("You win!")
        break
chat_socket.close()

    
    
```
