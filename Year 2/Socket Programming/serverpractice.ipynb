The following is a client program for an asymmetric guess-the-number game
where a server generates a random number from 1 to 100 and a client tries to
guess  it within  5  tries.  After each  incorrect guess,  the  server returns  whether 
the guess is greater than or less than the required number:

```python
import socket
import random
listen_socket = socket.socket()
listen_socket.bind(("127.0.0.1", 6003))
listen_socket.listen()

chat_socket, addr = listen_socket.accept()
# print(addr)
play_count = 0
target = random.randint(1, 100)
chat_socket.sendall(b"guess\n")
while True:
    data = b""
    while b"\n" not in data:
        data += chat_socket.recv(1024)
    try:
        msg = int(data.split(b"\n")[0].decode())
    except:
        continue
    # print(msg)
    if play_count == 5:
        chat_socket.sendall(b"over\n")
        break
    if msg == target:
        chat_socket.sendall(b"win\n")
        break
    elif msg < target:
        chat_socket.sendall(b"low\n")
    elif msg > target:
        chat_socket.sendall(b"high\n")
    chat_socket.sendall(b"guess\n")
    play_count += 1

chat_socket.close()
listen_socket.close()
```
