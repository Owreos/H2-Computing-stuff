**Task 1**

Write Python code to create the tables `Book` and `Author` with the following fields in the database `library.db`.

```
Table: BOOK
. ID INTEGER AUTO INCREMENT (Primary Key)
. NAME TEXT
. AUTHORID INTEGER  

Note: AUTHORID is a Foreign Key that is link to the Primary key of Author table
```

```
Table: AUTHOR
. ID INTEGER AUTO INCREMENT (Primary Key)
. NAME TEXT
```


````python
import sqlite3
query_1 = """
CREATE TABLE Author(
ID INTEGER PRIMARY KEY AUTOINCREMENT, Name TEXT
)
"""
query_2 = """
CREATE TABLE Book(
ID INTEGER PRIMARY KEY AUTOINCREMENT, Name TEXT, AuthorID INTEGER, FOREIGN KEY (AuthorID) REFERENCES Author(ID)
)
"""

with sqlite3.connect("library.db") as conn:
    with conn:
        conn.execute(query_1)
        conn.execute(query_2)

````

**Task 2**

Write a program that read the file `books.txt` and store the name of book in the `Book` table and name of the author in the `Author` table.

If there are no author information, replace the author as "unknown".

````python
import sqlite3

with open("books.txt", "r") as f:
    cursor = f.readlines()
    print(cursor)
    data = [x.strip().replace('"', "").split(" by ") for x in cursor if " by " in x]
    print(data)

query_check_author = "SELECT ID FROM Author WHERE Name = ?"
query_insert_book = "INSERT INTO Book(Name, AuthorID) VALUES(?, (SELECT ID FROM Author WHERE Name = ?))"
query_insert_author = "INSERT INTO Author(Name) VALUES(?)"

with sqlite3.connect("library.db") as conn:
    conn.execute("PRAGMA foreign_keys = ON")
    cur = conn.cursor()

    for book, author in data:
        cur.execute(query_check_author, (author,))
        row = cur.fetchone()
        if row is None:
            cur.execute(query_insert_author, (author,))
        cur.execute(query_insert_book, (book, author))
````
