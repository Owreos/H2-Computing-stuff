
## CP10 - Stack 

**Stack** is a Last in First Out (LIFO) structure in which access is completely restricted to just one end – this end is known as top.

### Operations
The basic operations of a stack is to add and remove item from its top. 
* **push()**: Add an item into the stack
* **pop()**: Remove item from the stack

Other supporting functions to be added are:
* **isEmpty()**: Returns true if the stack is empty
* **size()**: Returns the size of the stack
* **peek()**: Peek at the topmost item without removing it

### Exercise 1

Complete the `ArrayStack` class implementation using **Python list**:
* Initialize an empty list in initializer method
* Code `push()` and `pop()` functions to implement basic oprations of a stack

```python
class ArrayStack:
   def __init__(self):
      self.stack = [];
   def push(self, item):
      self.stack.append(item)
   def pop(self):
      return self.stack.pop(-1)
   def __str__(self):
      return ", ".join(self.stack)

stack = ArrayStack()
stack.push('a')
stack.push('b')
stack.push('c')
print(stack)
print(stack.pop())
print(stack.pop())
print(stack.pop() if stack else None)
```

Complete the `ArrayStack2` class implementation:
* It inherits from `ArrayStack` class
* Code the supplementry functions `size()`, `is_empty()`, `peek()`

```python
class ArrayStack2(ArrayStack):
    def init(self):
        super().__init__()
    
    def size(self):
        return len(self.stack)
    
    def is_empty(self):
        return not bool(self.stack)
    
    def peek(self):
        if not self.is_empty():
            return self.stack[-1]

    
stack = ArrayStack2()
stack.push('a')
stack.push('b')
print(stack.size())
stack.pop()
print(stack.peek())
stack.pop()
print(stack.is_empty())
print(stack.peek())
```

Complete the `ArrayStack3` class implementation:
* The stack has a size of 12
* Initialize a list in initializer method with a size of 12
* Code `push()`, `pop()`, `peek()`, `isEmpty()` and `size()` methods to implement basic operations of a stack.


```python
class ArrayStack3():
    def __init__(self):
        self._items = [None] * 12
        self.top = -1
        self.arraysize = 12
    
    def push(self, item):
        if self.top == 12 - 1:
            raise Exception("stack overflow")
        self.top += 1
        self._items[self.top] = item

    def pop(self):
        if self.is_empty():
            raise Exception("stack underflow")
        popped = self._items[self.top]
        self._items[self.top] = None
        self.top -= 1
        return popped
        
    def peek(self):
        if self.is_empty():
            raise Exception("stack is empty")
        return self._items[self.top]
    
    def is_empty(self):
        return self.top == -1
    
    def size(self):
        return self.top + 1
        
    
stack = ArrayStack3()
stack.push('a')
stack.push('b')
stack.push('c')
print(stack._items)
print(stack.pop())
print(stack.pop())
print(stack.pop() if stack else None)
```

### Exercise 4

Complete the Stack class implementation using the singly linked list which you have implemented previously:

Define the Node and LinkedList class and The LinkedListStack class

```python
class Node:
   def __init__(self, value = None, next = None):
      self.value = value
      self.next = next

class LinkedList():
     def __init__(self, head = None):
         self._head = head
         
class LinkedListStack(): # for this access from start is most efficent
     def __init__(self):
         self._items = LinkedList()

     def push(self, node):
          node.next = self._items._head
          self._items._head = node

     def pop(self):
         if self._items._head == None:
             raise Exception("stack underflow")
         popped = self._items._head
         self._items._head = self._items._head.next
         return popped.value
     
     def is_empty(self):
        return self._items._head is None
     def peek(self):
        return self._items._head.value
     def size(self):
        probe = self._items._head
        count = 0
        while probe:
            count += 1
            probe = probe.next

         
stack = LinkedListStack()
stack.push(Node('a'))
stack.push(Node('b'))
stack.push(Node('c'))
print(stack.pop())
print(stack.pop())
print(stack.pop() if stack else None)
    
```
