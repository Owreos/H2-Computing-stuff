## Queue ADT

**Queue** is a First in First Out (FIFO) structure in which access is completely restricted to just one end – this end is known as top.

### Operations
The basic operations of a queue areto enqueue and dequeue item from its top. 
* **enqueue**: Add an item at the end of the queue.
* **dequeue**: Remove and return the first item of the queue, if the queue is not empty

Other supporting functions to be added are:
* **isEmpty()**: Returns true if the queue is empty, otherwise false
* **size()**: Returns the number of items in the queue
* **peek()**: return the first item of the queue without removing it, if the queue is not empty

```python
# normal queue implementation
class Queue():
    def __init__(self):
        self.queue = []

    def enqueue(self, item):
        self.queue.append(item)
    
    def dequeue(self):
        return self.queue.pop(0)
    
    def isEmpty(self):
        return self.queue is not None
    
    def size(self):
        return self.queue.len()

    def peek(self):
        return self.queue[0]
    
testqueue = Queue()
print(testqueue.isEmpty())
testqueue.enqueue("A")
print(testqueue.dequeue())

```

```python
class ArrayQueue():
    def __init__(self, size):
        self.queue = [None]*size
        self.top = -1
        self.size = 0

    def enqueue(self, item):
        if self.top == len(self.queue) - 1:
            raise Exception("stack overflow")
        self.top += 1
        self.size += 1
        self.queue[self.top] = item
    
    def dequeue(self):
        if self.top == -1:
            raise Exception("stack underflow")
        popped = self.queue[self.top]
        self.queue[self.top] = None
        self.top -= 1
        self.size -= 1
        return popped
    
    def isEmpty(self):
        return self.size == 0
    
    def size(self):
        return self.size

    def peek(self):
        return self.queue[self.top]
# normal queue implementation0]
    
testqueue = ArrayQueue(12)
print(testqueue.isEmpty())
testqueue.enqueue("A")
print(testqueue.dequeue())
```

#### Implement ArrayQueue2 with the following :
* Maintain a variable `front` that points to the item at the front of the queue. Starts at 0 and advances as items dequeued.
* Rear variable points to the last item at position n-1, where n is the number of items in queue
* The items will be shifted to the start of the queue when the rear pointer is about to run off the end

```python
class ArrayQueue():
    def __init__(self):
        self.queue = [None] * 12
        self.rear = -1
        self.front = 0
        self.count = 0

    def enqueue(self, item):
        if self.rear + 1 == len(self.queue):
            if self.front == 0:
                raise Exception("queue overflow")
            self.shift_helper()
            self.front = 0
            self.rear = self.count - 1
        self.rear += 1
        self.queue[self.rear] = item
        self.count += 1

    def shift_helper(self):
        for i in range(self.count):
             self.queue[i] = self.queue[i + self.front]
             self.queue[i + self.front] = None
    
    def dequeue(self):
        if self.isEmpty():
            raise Exception("queue underflow vro")
        popped = self.queue[self.front]
        self.queue[self.front] = None
        self.front += 1
        self.count -= 1
        return popped
    
    def isEmpty(self):
        return self.count == 0
    
    def size(self):
        return self.count

    def peek(self):
        return self.queue[self.front]
# normal queue implementation0]
    
testqueue = ArrayQueue(12)
print(testqueue.isEmpty())
testqueue.enqueue("A")
print(testqueue.dequeue())
```
