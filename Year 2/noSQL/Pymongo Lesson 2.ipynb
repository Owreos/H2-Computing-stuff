#### Find all documents in the collection

* Using find() to list `all` documents in a collection.

    ```
    db = client.mydatabase
    for document in db.mycollection.find():
        print(document)
    ```
 
* If you are only interested in the first match for a given query, or if you know the database only contains one matching document, you can use `find_one()`.

#### Find documents in the collection with conditions/filters
Projection in MongoDB follows some basic rules:

- The `_id` field is always included unless explicitly excluded. 
- Specifying a field for inclusion implicitly excludes all other fields except the `_id` field.
    
    ```
    db = client.mydatabase
    query = {"age": {"$gt":20}}  ## filter for age greater than 20
    projections ={"_id":0}
    
    result = db.books.find(query, projections)
    for x in result:
        print(x)   
    ```
