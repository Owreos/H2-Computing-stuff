#### Update one document using db.collection.update_one()

The update_one() method updates the first document that satisfies the query in a collection.

```
    myquery = {"address": "Valley 345"}
    newval = {"$set":{"address":"Canyon 123"}}
    
    db = client.mydatabase
    result = db.customers.update_one(myquery, newval)
   
    for document in db.customers.find():
        print(document)
```
    

#### Update many documents using db.collection.update_many()

The update_many() method updates all documents that satisfy the query in a collection.

```
    db = client.mydatabase
    myquery = {"name": "Minnies"}
    newval = {"$set":{"name":"Minnie"}}
    
    result = db.customers.update_many(myquery, newval)
    print(f"No of documents updated : {result.modified_count}")

```

#### Task 4
Find books with title ``containing`` the word **Hadoop**. 

Hint: use ``"$regex"``

Display the result after the update as shown:

```
ID                   Title                                              ISBN                
186                  Hadoop in Action                                   1935182196          
566                  Hadoop in Practice                                 1617290238          
772                  Hadoop in Practice, Second Edition                 1617292222                
```

````python
client = MongoClient("127.0.0.1", 27017)
db = client.bookdb
query = {"title": {"$regex": "Hadoop"}}
projection = {"_id" : 1, "title": 1, "isbn": 1}
res = db.book.find(query, projection)
print(f"{'ID':40}{'Title':40}{'ISBN':20}")
for i in res:
    id = i.get("_id", "")
    title = i.get("title", "")
    isbn = i.get("isbn", "")
    print(f"{id:<40}{title:40}{isbn:20}")
````

#### Task 5

Update the book ``Hadoop in Action`` to ``Hadoop in Action eBook``.

Display the result after the update as shown:

```
ID                   Title                                              ISBN                
186                  Hadoop in Action eBook                             1935182196          
566                  Hadoop in Practice                                 1617290238          
772                  Hadoop in Practice, Second Edition                 1617292222       
```

````python
client = MongoClient("127.0.0.1", 27017)
db = client.bookdb
query = {"title": {"$regex": "Hadoop"}}
projection = {"_id" : 1, "title": 1, "isbn": 1}

replace = {"$set": {"title": "Hadoop in Action eBook"}}

replaced = db.book.update_many(query, replace)

print(replaced.upserted_id)

res = db.book.find(query, projection)
print(f"{'ID':40}{'Title':40}{'ISBN':20}")
for i in res:
    id = i.get("_id", "")
    title = i.get("title", "")
    isbn = i.get("isbn", "")
    print(f"{id:<40}{title:40}{isbn:20}")
````

#### Task 6

Update the status of documents with "PUBLISH" status to "PUBLISHED".

Print the number of the documents with status = "PUBLISHED" in the database after the update.



```
Status:PUBLISHED Count:363
     
```

````python
client = MongoClient("127.0.0.1", 27017)
db = client.bookdb
query = {"status": "PUBLISH"}
replace = {"$set": {"status": "PUBLISHED"}}
db.book.update_many(query, replace)
countquery = {"status": "PUBLISHED"}
countres = db.book.count_documents(countquery)
print(f"Status: published  Count: {countres}")
````
